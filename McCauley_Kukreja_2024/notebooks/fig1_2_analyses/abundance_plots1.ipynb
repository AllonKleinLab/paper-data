# Calculate abundances

```python
# function to percent cell type abundance:
def rep_abundance(adata_rep, condition_list, cell_state_order, celltype_obs = 'cell_states' ):
    cell_abundance_dict = {}
    for conditions in condition_list:
        abundance =  pd.Series(Counter(pd.Series(np.array(adata_rep[adata_rep.obs.condition == conditions].obs[celltype_obs]))),
                              dtype = 'float64')
        pd_temp = 100*abundance/abundance.sum() 
        cell_abundance_dict[conditions] = pd_temp

    cell_abundances = pd.DataFrame(cell_abundance_dict).T
    cell_abundances = cell_abundances[cell_state_order]
    cell_abundances = cell_abundances.loc[condition_list]
    
    return cell_abundances
```

# Add p-values using Fisher exact test

## Function to find absolute abundances

```python
# get absolute abundances (i.e. cell numbers and not percentages):

def rep_abundance_abs(adata_rep, condition_list, cell_state_order, celltype_obs = 'cell_states' ):
    cell_abundance_dict = {}
    for conditions in condition_list:
        abundance =  pd.Series(Counter(pd.Series(np.array(adata_rep[adata_rep.obs.condition == conditions].obs[celltype_obs]))),
                              dtype = 'float64')
        pd_temp = abundance
        cell_abundance_dict[conditions] = pd_temp

    cell_abundances = pd.DataFrame(cell_abundance_dict).T
    cell_abundances = cell_abundances[cell_state_order]
    cell_abundances = cell_abundances.loc[condition_list]
    
    return cell_abundances
```

## Function to find p-values

```python
def pval_fisher(abundances):
    abundances = abundances.dropna()
    pval_abundance = pd.DataFrame(np.zeros(abundances.shape))
    pval_abundance.index = abundances.index
    pval_abundance.columns = abundances.columns
    
    for i,conditions in enumerate(abundances.index):
        for j,states in enumerate(abundances.columns):
            a1=abundances[states]['None']
            b1=abundances.loc['None'].sum() - abundances[states]['None']

            a2=abundances[states][conditions]
            b2=abundances.loc[conditions].sum() - abundances[states][conditions]
            
            pval_abundance.loc[conditions,states] = ss.fisher_exact(np.array([[a1,a2],[b1,b2]]),
                                                                    alternative='two-sided')[1]
            
        
    fdr_abundance_temp= pval_abundance.drop('None')
    fdr_abundance = stats.multitest.multipletests(np.array(fdr_abundance_temp).flatten(),  method='fdr_bh')[1]
    fdr_abundance = pd.DataFrame(fdr_abundance.reshape(fdr_abundance_temp.shape))
    fdr_abundance.index = fdr_abundance_temp.index
    fdr_abundance.columns = fdr_abundance_temp.columns
    
    return fdr_abundance
```

# Combine p-values for supplementary table

```python
def pval_fisher_raw(abundances):
    abundances = abundances.dropna()
    pval_abundance = pd.DataFrame(np.zeros(abundances.shape))
    pval_abundance.index = abundances.index
    pval_abundance.columns = abundances.columns
    
    for i,conditions in enumerate(abundances.index):
        for j,states in enumerate(abundances.columns):
            a1=abundances[states]['None']
            b1=abundances.loc['None'].sum() - abundances[states]['None']

            a2=abundances[states][conditions]
            b2=abundances.loc[conditions].sum() - abundances[states][conditions]
            
            pval_abundance.loc[conditions,states] = ss.fisher_exact(np.array([[a1,a2],[b1,b2]]),
                                                                    alternative='two-sided')[1]
    return pval_abundance    
```

```python
add_rare_subsets4 = pd.DataFrame([cell_abundances4_abs['Rare']/3,cell_abundances4_abs['Rare']/3,cell_abundances4_abs['Rare']/3]).T
add_rare_subsets4.columns = ['Ionocyte','Tuft','PNEC']
cell_abundances4_abs = pd.concat([cell_abundances4_abs,add_rare_subsets4], axis=1)
cell_abundances4_abs = cell_abundances4_abs.drop(['Rare'], axis=1)

add_rare_subsets5 = pd.DataFrame([cell_abundances5_abs['Rare']/3,cell_abundances5_abs['Rare']/3,cell_abundances5_abs['Rare']/3]).T
add_rare_subsets5.columns = ['Ionocyte','Tuft','PNEC']
cell_abundances5_abs = pd.concat([cell_abundances5_abs,add_rare_subsets5], axis=1)
cell_abundances5_abs = cell_abundances5_abs.drop(['Rare'], axis=1)

add_rare_subsets6 = pd.DataFrame([cell_abundances6_abs['Rare']/3,cell_abundances6_abs['Rare']/3,cell_abundances6_abs['Rare']/3]).T
add_rare_subsets6.columns = ['Ionocyte','Tuft','PNEC']
cell_abundances6_abs = pd.concat([cell_abundances6_abs,add_rare_subsets6], axis=1)
cell_abundances6_abs = cell_abundances6_abs.drop(['Rare'], axis=1)
```

```python
pval_raw1 = pval_fisher_raw(cell_abundances1_abs)
pval_raw1 = pval_raw1.iloc[1:,:]

pval_raw2 = pval_fisher_raw(cell_abundances2_abs)
pval_raw2 = pval_raw2.iloc[1:,:]

pval_raw3 = pval_fisher_raw(cell_abundances3_abs)
pval_raw3 = pval_raw3.iloc[1:,:]

pval_raw4 = pval_fisher_raw(cell_abundances4_abs)
pval_raw4 = pval_raw4.iloc[1:,:]

pval_raw5 = pval_fisher_raw(cell_abundances5_abs)
pval_raw5 = pval_raw5.iloc[1:,:]

pval_raw6 = pval_fisher_raw(cell_abundances6_abs)
pval_raw6 = pval_raw6.iloc[1:,:]
```

```python
# make dictionary of replicates: 
replicate_dict = {}
for condition in np.array(condition_names)[1:]:
    temp = np.array([False, False, False, False, False, False])
    if condition in pval_raw1.index:
        temp[0] = True
    if condition in pval_raw2.index:
        temp[1] = True
    if condition in pval_raw3.index:
        temp[2] = True
    if condition in pval_raw4.index:
        temp[3] = True
    if condition in pval_raw5.index:
        temp[4] = True
    if condition in pval_raw6.index:
        temp[5] = True

    replicate_dict[condition] = temp
```

```python
# make array of pvalue dataframes:
pvals_individual = np.array([pval_raw1, pval_raw2,pval_raw3,pval_raw4, pval_raw5, pval_raw6]) 

# initialize dataframe with combined p-values:
pval_combined = np.zeros((17,7))
pval_combined = pd.DataFrame(pval_combined)
pval_combined.index = np.array(condition_names)[1:]
pval_combined.columns =  pval_raw1.columns

for condition in np.array(condition_names)[1:]:
    pvals_to_consider = pvals_individual[replicate_dict[condition]] # select replicates: this step 
        #is required because not all conditions have 3 replicates so we need to restrict to
        #the replicates for that are applicate for that condition
    for states in pval_raw1.columns:
        pval_state = []
        for items in pvals_to_consider:
            pval_state = pval_state+[items.loc[condition][states]]
        pval_combined.loc[condition,states] = ss.combine_pvalues(pval_state, method='fisher')[1]

 
```
