# What this notebook does

This notebook plots expression of top 20 genes of TGFB1 enriched programs in abberant cells and rest of the cells. It contains the code for generating figure 5f of McCauley/Kukreja et al. 2024 manuscript

# Set up 

```python
import numpy as np, pandas as pd, matplotlib.pyplot as pl, scipy.sparse as ssp
import scanpy as sc,scanpy.external as sce
import sys
sys.path.append('../../helper_functions/')
import helper_functions_py3 as hf

sc.settings.verbosity = 2  # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.settings.set_figure_params(dpi=100)  # low dpi (dots per inch) yields small inline figures

import random
import scipy.stats  as ss
import statsmodels.stats.multitest as multitest
from matplotlib.patches import Rectangle
import seaborn as sn
```

# Load data

```python
# load data from all diseases:
# update this with the path where your anndata objects are stored !!
anndata_path = '../../data/anndata_objects/'

# IPF1, GEO ID: GSE135893
adata_ipf1 = sc.read_h5ad(anndata_path + 'adata_ipf1_epithelial_processed.h5ad')

# IPF 2, GEO ID: GSE136831
adata_ipf2 = sc.read_h5ad(anndata_path + 'adata_ipf2_epithelial_processed.h5ad')

# COPD, GEO ID: GSE136831
adata_copd = sc.read_h5ad(anndata_path + 'adata_copd_epithelial_processed.h5ad')

# COVID, GEO ID: GSE171524
adata_covid = sc.read_h5ad(anndata_path +'adata_covid_epithelial_processed.h5ad')
```

```python
# load gene program data, also called the spectra matrix:

# this is gene x programs matrix with a weight for every gene 
# for each of the gene expression programs

cnmf_dir = '../../data/cnmf/'
spectra = pd.read_csv(cnmf_dir+'program_weights.csv', sep=',', index_col=0).T
```

```python
# subset genes to only include the ones in disease dataset:

# ipf1:
gene_subset_mask_ipf1 = np.in1d(spectra.columns, adata_ipf1.var_names)
spectra_subset_ipf1 = spectra[spectra.columns[gene_subset_mask_ipf1]]

# ipf2:
gene_subset_mask_ipf2 = np.in1d(spectra.columns, adata_ipf2.var_names)
spectra_subset_ipf2 = spectra[spectra.columns[gene_subset_mask_ipf2]]

# copd:
gene_subset_mask_copd = np.in1d(spectra.columns, adata_copd.var_names)
spectra_subset_copd = spectra[spectra.columns[gene_subset_mask_copd]]

# covid:
gene_subset_mask_covid = np.in1d(spectra.columns, adata_covid.var_names)
spectra_subset_covid = spectra[spectra.columns[gene_subset_mask_covid]]
```

```python
# subset genes that are common in all datasets:
gene_list_common = np.intersect1d(np.intersect1d(np.intersect1d(spectra_subset_ipf1.columns, spectra_subset_ipf2.columns),
                              spectra_subset_copd.columns), spectra_subset_covid.columns)


spectra_updated = spectra[gene_list_common]
```

# Make heatmap of TGFB specific gene expression

```python
# top 20 genes for programs: TGFB1-1, TGFB1-2
num_genes = 20

tgfb_prog1 = spectra_updated.loc['TGFB1-1'].sort_values(ascending = False).index[:num_genes]
tgfb_prog2 = spectra_updated.loc['TGFB1-2'].sort_values(ascending = False).index[:num_genes]
ifng_prog1 = spectra_updated.loc['IFNG-1'].sort_values(ascending = False).index[:num_genes]
ifng_prog3 = spectra_updated.loc['IFNG-3'].sort_values(ascending = False).index[:num_genes]
```

```python
# making heatmap for expression of top genes in the program TGFB-1:
# this includes expression in control cells, non-aberrant disease cells and aberrant disease cells
# expression is calculated for the 4 datasets used

tgfb_prog1_heatmap = (np.vstack([adata_ipf1[:,tgfb_prog1][adata_ipf1.obs.Diagnosis == 'Control'].X.mean(axis=0),
                                 adata_ipf2[:,tgfb_prog1][adata_ipf2.obs['"Disease_Identity"'] == '"Control"'].X.mean(axis=0),
                                 adata_copd[:,tgfb_prog1][adata_copd.obs['"Disease_Identity"'] == '"Control"'].X.mean(axis=0),
                                 adata_covid[:,tgfb_prog1][adata_covid.obs.disease__ontology_label == 'normal'].X.mean(axis=0),
                                 
                                 adata_ipf1[:,tgfb_prog1][(adata_ipf1.obs.Diagnosis == 'IPF')&(adata_ipf1.obs.celltype != 'KRT5-/KRT17+')].X.mean(axis=0),
                                 adata_ipf2[:,tgfb_prog1][(adata_ipf2.obs['"Disease_Identity"'] == '"IPF"')&(adata_ipf2.obs['"Manuscript_Identity"'] != 'Aberrant_Basaloid')].X.mean(axis=0),
                                 adata_copd[:,tgfb_prog1][(adata_copd.obs['"Disease_Identity"'] == '"COPD"')&(adata_copd.obs['"Manuscript_Identity"'] != 'Aberrant_Basaloid')].X.mean(axis=0),
                                 adata_covid[:,tgfb_prog1][(adata_covid.obs.disease__ontology_label == 'COVID-19')&(adata_covid.obs.cell_type_fine != 'ECM-high epithelial')].X.mean(axis=0),
                                 
                                 adata_ipf1[:,tgfb_prog1][(adata_ipf1.obs.Diagnosis == 'IPF')&(adata_ipf1.obs.celltype == 'KRT5-/KRT17+')].X.mean(axis=0),
                                 adata_ipf2[:,tgfb_prog1][(adata_ipf2.obs['"Disease_Identity"'] == '"IPF"')&(adata_ipf2.obs['"Manuscript_Identity"'] == 'Aberrant_Basaloid')].X.mean(axis=0),
                                 adata_copd[:,tgfb_prog1][(adata_copd.obs['"Disease_Identity"'] == '"COPD"')&(adata_copd.obs['"Manuscript_Identity"'] == 'Aberrant_Basaloid')].X.mean(axis=0),
                                 adata_covid[:,tgfb_prog1][(adata_covid.obs.disease__ontology_label == 'COVID-19')&(adata_covid.obs.cell_type_fine == 'ECM-high epithelial')].X.mean(axis=0)
                                 
                                ]))
```

```python
# making heatmap for expression of top genes in the program TGFB-2:
# this includes expression in control cells, non-aberrant disease cells and aberrant disease cells
# expression is calculated for the 4 datasets used

tgfb_prog2_heatmap = (np.vstack([adata_ipf1[:,tgfb_prog2][adata_ipf1.obs.Diagnosis == 'Control'].X.mean(axis=0),
                                 adata_ipf2[:,tgfb_prog2][adata_ipf2.obs['"Disease_Identity"'] == '"Control"'].X.mean(axis=0),
                                 adata_copd[:,tgfb_prog2][adata_copd.obs['"Disease_Identity"'] == '"Control"'].X.mean(axis=0),
                                 adata_covid[:,tgfb_prog2][adata_covid.obs.disease__ontology_label == 'normal'].X.mean(axis=0),
                                 
                                 adata_ipf1[:,tgfb_prog2][(adata_ipf1.obs.Diagnosis == 'IPF')&(adata_ipf1.obs.celltype != 'KRT5-/KRT17+')].X.mean(axis=0),
                                 adata_ipf2[:,tgfb_prog2][(adata_ipf2.obs['"Disease_Identity"'] == '"IPF"')&(adata_ipf2.obs['"Manuscript_Identity"'] != 'Aberrant_Basaloid')].X.mean(axis=0),
                                 adata_copd[:,tgfb_prog2][(adata_copd.obs['"Disease_Identity"'] == '"COPD"')&(adata_copd.obs['"Manuscript_Identity"'] != 'Aberrant_Basaloid')].X.mean(axis=0),
                                 adata_covid[:,tgfb_prog2][(adata_covid.obs.disease__ontology_label == 'COVID-19')&(adata_covid.obs.cell_type_fine != 'ECM-high epithelial')].X.mean(axis=0),
                                 
                                 adata_ipf1[:,tgfb_prog2][(adata_ipf1.obs.Diagnosis == 'IPF')&(adata_ipf1.obs.celltype == 'KRT5-/KRT17+')].X.mean(axis=0),
                                 adata_ipf2[:,tgfb_prog2][(adata_ipf2.obs['"Disease_Identity"'] == '"IPF"')&(adata_ipf2.obs['"Manuscript_Identity"'] == 'Aberrant_Basaloid')].X.mean(axis=0),
                                 adata_copd[:,tgfb_prog2][(adata_copd.obs['"Disease_Identity"'] == '"COPD"')&(adata_copd.obs['"Manuscript_Identity"'] == 'Aberrant_Basaloid')].X.mean(axis=0),
                                 adata_covid[:,tgfb_prog2][(adata_covid.obs.disease__ontology_label == 'COVID-19')&(adata_covid.obs.cell_type_fine == 'ECM-high epithelial')].X.mean(axis=0)
                                 
                                ]))
```

```python
# plot data for expression of TGFB1-1 genes across the 4 disease datasets 
# and 3 different groups of cells: control cells, non-aberrant disease cells and aberrant disease cells

pl.figure(figsize = (6,3))
sn.heatmap(tgfb_prog1_heatmap, cmap = 'RdBu_r', vmin = -3, vmax = 3, 
          yticklabels=np.arange(tgfb_prog1_heatmap.shape[0]), xticklabels = tgfb_prog1)
bla = pl.xticks(fontstyle = 'italic')
pl.axhline(y=4, color = 'white', lw=2)
pl.axhline(y=8, color = 'white', lw=2)

pl.savefig("../../figures/Fig5F_left_tgfb1-1_genes.pdf", bbox_inches='tight', dpi = 300)
```
