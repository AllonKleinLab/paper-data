# What this notebook does

1. This notebook combines the heatmaps of fold change program expression in the 4 disease datasets (see Fig. 5B for citation of the 4 disease datasets).

    * Fold change program expression for the 4 datasets were calculated in the notebooks: prog_expr_ipf1.ipynb, prog_expr_ipf2.ipynb, prog_expr_copd.ipynb, prog_expr_covid.ipynb.

    * This notebook just combines the heatmap for all disease datasets for the purpose of plotting Figure 5C.
    
2. This notebook also plots fold change of immune signaling programs in covid disease datasets for Figure 5D. 
    * The fold changes were calculated in the notebook: prog_exp_covid.ipynb. The notebook simply plots these as box plots.

# Set up

```python
import numpy as np, pandas as pd, matplotlib.pyplot as pl, scipy.sparse as ssp
import scanpy as sc,scanpy.external as sce
import sys
sys.path.append('../../helper_functions/')
import helper_functions_py3 as hf

sc.settings.verbosity = 2  # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.settings.set_figure_params(dpi=100)  # low dpi (dots per inch) yields small inline figures

import random
import scipy.stats  as ss
import statsmodels.stats.multitest as multitest
from matplotlib.patches import Rectangle
import seaborn as sn
```

# Load data

```python
copd_data = pd.read_table('../../data/csv_txt_files/copd_heatmap_bootstrapped.csv', sep = ',', index_col = 0)
covid_data = pd.read_table('../../data/csv_txt_files/covid_heatmap_bootstrapped.csv', sep = ',', index_col = 0)
ipf1_data = pd.read_table('../../data/csv_txt_files/ipf1_heatmap_bootstrapped.csv', sep = ',', index_col = 0)
ipf2_data = pd.read_table('../../data/csv_txt_files/ipf2_heatmap_bootstrapped.csv', sep = ',', index_col = 0)

```

# Make heatmap of all data combined

```python
copd_index = np.zeros(copd_data.shape[0], dtype = object)
for i, ix in enumerate(copd_data.index):
    copd_index[i] = 'COPD_'+ix
copd_data.index = copd_index

covid_index = np.zeros(covid_data.shape[0], dtype = object)
for i, ix in enumerate(covid_data.index):
    covid_index[i] = 'Covid_'+ix
covid_data.index = covid_index
    
ipf1_index = np.zeros(ipf1_data.shape[0], dtype = object)
for i, ix in enumerate(ipf1_data.index):
    ipf1_index[i] = 'IPF1_'+ix
ipf1_data.index = ipf1_index
    
ipf2_index = np.zeros(ipf2_data.shape[0], dtype = object)
for i, ix in enumerate(ipf2_data.index):
    ipf2_index[i] = 'IPF2_'+ix
ipf2_data.index = ipf2_index
```

```python
all_data = pd.concat([ ipf1_data,ipf2_data, copd_data, covid_data])
```

```python
all_data.index
```

```python


index_new = ['IPF1_AT1', 'IPF2_ATI','COPD_ATI',  'Covid_AT1', 
             'IPF1_AT2','IPF1_Transitional AT2', 'IPF2_ATII','COPD_ATII', 'Covid_AT2',
             'IPF1_Basal', 'IPF2_Basal', 'COPD_Basal','Covid_Airway basal', 
             'IPF1_MUC5B+', 'IPF1_SCGB3A2+',
       'IPF1_SCGB3A2+ SCGB1A1+','IPF2_Club','COPD_Club',  'Covid_Airway club',
             'IPF1_MUC5AC+ High','IPF2_Goblet', 'COPD_Goblet','Covid_Airway goblet',
             'IPF1_Ciliated','IPF1_Differentiating Ciliated', 'IPF2_Ciliated', 'COPD_Ciliated', 'Covid_Airway ciliated',
             'IPF2_PNEC',  'IPF2_Ionocyte',  'COPD_PNEC','Covid_Tuft-like',
             'IPF1_Proliferating Epithelial Cells', 'Covid_Cycling epithelial',
             'IPF1_KRT5-/KRT17+', 
             'IPF2_Aberrant_Basaloid', 'COPD_Aberrant_Basaloid','Covid_ECM-high epithelial']
```

```python
all_data = all_data.loc[index_new]
```

```python
all_data.head()
```

## Plot heatmap

```python
# subset signifant differences and fc > 0

# initialize plot
pl.figure(figsize = (all_data.shape[1]/3,all_data.shape[0]/3))
pl.axhline(4, color='white', lw=2)
pl.axhline(9, color='white', lw=2)
pl.axhline(13, color='white', lw=2)
pl.axhline(19, color='white', lw=2)
pl.axhline(23, color='white', lw=2)
pl.axhline(28, color='white', lw=2)
pl.axhline(32, color='white', lw=2)
pl.axhline(34, color='white', lw=2)

# plot grey region
sn.heatmap(-all_data.isnull(),  xticklabels=False, yticklabels = False,
              cmap = 'Greys', vmax = 1, vmin = -0.3, cbar=False, rasterized = True)

# plot FC
g =sn.heatmap(all_data, cmap = 'Reds' , vmax = 2, vmin = 0, square = True,
          yticklabels=all_data.index, 
              cbar=False, rasterized = True)

pl.xticks(rotation = 90, fontstyle = 'italic', fontsize = 15)
pl.yticks(np.arange(all_data.shape[0])+0.5,labels = [], fontstyle = 'italic')

# add box around fold change greater than 1
add_boxes_bool = np.array((all_data>1)&((1-np.isnan(all_data)).astype(bool)))
r,c = np.where(add_boxes_bool)
for i in range(r.shape[0]):
    g.add_patch(Rectangle((c[i], r[i]), 1, 1, fill=False, edgecolor='black', lw=2))
    
pl.xticks(rotation = 90)
# linewidths=1, linecolor='white'
pl.savefig("../../figures/Fig5C_all_disease_combined_heatmap_bootstrapped.pdf", bbox_inches='tight', dpi = 300)
```

# Expression of interferon genes in Covid samples

```python
pval_covid = pd.read_table('../../data/csv_txt_files/pval_covid_bootstrapped.csv', sep = ',')
fc_covid = pd.read_table('../../data/csv_txt_files/fc_covid_bootstrapped.csv', sep = ',', index_col=0)
```

```python
covid_ifna = pd.DataFrame(np.array(fc_covid['IFNA']))
covid_ifna['Program'] = 'IFNA'*np.ones(fc_covid['IFNA'].shape[0], dtype = 'object')


covid_ifng1 = pd.DataFrame(np.array(fc_covid['IFNG-1']))
covid_ifng1['Program'] = 'IFNG-1'*np.ones(fc_covid['IFNG-1'].shape[0], dtype = 'object')

covid_ifng2 = pd.DataFrame(np.array(fc_covid['IFNG-2']))
covid_ifng2['Program'] = 'IFNG-2'*np.ones(fc_covid['IFNG-2'].shape[0], dtype = 'object')

covid_ifng3 = pd.DataFrame(np.array(fc_covid['IFNG-3']))
covid_ifng3['Program'] = 'IFNG-3'*np.ones(fc_covid['IFNG-3'].shape[0], dtype = 'object')


covid_il13 = pd.DataFrame(np.array(fc_covid['IL13']))
covid_il13['Program'] = 'IL13'*np.ones(fc_covid['IL13'].shape[0], dtype = 'object')
```

```python
covid_ifna_plot = pd.concat([covid_ifng3,covid_il13,covid_ifng2,covid_ifna,covid_ifng1])
covid_ifna_plot.columns = ['FC', 'Program']
```

```python
pl.figure(figsize = (3,3))
sn.boxplot(x = 'Program', y = 'FC', data = covid_ifna_plot, color = '#A9A9A9', width = 0.6)
sn.stripplot(x = 'Program', y = 'FC', data = covid_ifna_plot, color = 'black')

pl.grid(False)

pl.yticks([-2,-1,0,1,2,3,4], fontname = 'Arial')
pl.xticks(np.arange(5), label = ['IFNG-3','IL13', 'IFNG-2','IFNA','IFNG-1' ], rotation = 90, style = 'italic', fontname = 'Arial')
#pl.axhline(y=1, ls='dashed', color = 'Red')
sn.despine()
#pl.ylim((-1,1))
pl.xlabel(None)
pl.ylabel(None)

pl.axhline(y=0, ls = 'dashed', color = 'Red')

pl.savefig("../../figures/Fig5d_covid_fc_bootstrapped.pdf", bbox_inches='tight', dpi = 300)
```
