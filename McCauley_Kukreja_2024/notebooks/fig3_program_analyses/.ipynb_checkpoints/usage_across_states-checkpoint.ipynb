# What this notebook does:

This notebook calculates:
* Usages of control versus perturbation programs in the following conditions: Untreated, BMP4, IFNA, CHIR, TGFB1 and IFNG for the cell states: basal, secretory, multiciliated and rare cells. These calculations generate Fig.3F of McCauley/Kukreja et. al. 2024 manuscript.
* It also calculates cell type specific expression of various programs in specific signaling conditions (Fig.3D and S5C).

```python
# load cnmf data:
cnmf_dir = '../../data/cnmf/'
spectra = pd.read_csv(cnmf_dir+'program_weights.csv', sep=',', index_col=0).T
usage_matrix = pd.read_csv(cnmf_dir+'program_usages.csv')
normalized_usage_matrix = usage_matrix.div(usage_matrix.sum(axis=1), axis=0)
```

# Balance of control vs. perturbed programs

```python
# combine club and goblet states into a single secretory state 
# and rare cell states into a single category

state_list_coarse = np.array(['Basal', 'Secretory','Multiciliated', 'Rare'])
state_coarse = np.array(adata.obs.cell_states)
for i,states in enumerate(state_coarse):
    if states == 'Goblet':
        state_coarse[i] = 'Secretory'
    if states == 'Club':
        state_coarse[i] = 'Secretory'
    if states == 'Ionocyte':
        state_coarse[i] = 'Rare'
    if states == 'Tuft':
        state_coarse[i] = 'Rare'
    if states == 'PNEC':
        state_coarse[i] = 'Rare'
adata.obs['state_coarse'] = state_coarse        
```

```python
# function to make dictionary of total usages of ctrl programs and 
# perturb programs for each cell type, given a particular condition

def gep_exp_dist(adata, treatment):

    adata_temp = adata[adata.obs.condition == treatment].copy()
    normalized_usage_matrix_subset = normalized_usage_matrix[np.array(adata.obs.condition == treatment)]

    condition_state = np.array(list(set(adata[adata.obs.condition == treatment].obs.state_coarse)))
    state_list_updated = state_list_coarse[np.in1d(state_list_coarse,condition_state)]

    state_usages_subset = {}

    for i,states in enumerate(state_list_updated):
        temp = normalized_usage_matrix_subset[np.array(adata_temp.obs.state_coarse == states)]
        temp_ctrl = temp.iloc[:,:9].sum(1)
        temp_pert = temp.iloc[:,9:].sum(1)
        df = pd.DataFrame([np.array(temp_ctrl), np.array(temp_pert)]).T
        df.columns = ['Control', 'Perturb']
        state_usages_subset[states] = df.copy()

    return state_usages_subset

```

```python
# make a dictionary with total usages of control and perturbed programs
# for every state and condition:

prog_dist = {}
for conditions in condition_list:
    prog_dist[conditions]= gep_exp_dist(adata,conditions)
    
```

# Expression of perturbation programs across cell types

```python
def gep_exp1(adata, treatment):

    adata_temp = adata[adata.obs.condition == treatment].copy()
    normalized_usage_matrix_subset = normalized_usage_matrix[np.array(adata.obs.condition == treatment)]
    
    condition_state = np.array(list(set(adata[adata.obs.condition == treatment].obs.cell_states)))
    state_list_updated = state_list[np.in1d(state_list,condition_state)]

    state_usages_subset = np.zeros((state_list_updated.shape[0],spectra.shape[0]))

    for i,states in enumerate(state_list_updated):
        state_usages_subset[i,:] = np.array(normalized_usage_matrix_subset[np.array(adata_temp.obs.cell_states == states)].mean())

    state_usages_subset_pd = pd.DataFrame(state_usages_subset)
    
    state_usages_subset_pd.columns = normalized_usage_matrix_subset.columns
    state_usages_subset_pd.index = state_list_updated
    

    pl.figure(figsize = (state_usages_subset_pd.shape[1]/2,state_usages_subset_pd.shape[0]/2))
    sn.heatmap(state_usages_subset_pd, cmap = 'Reds', yticklabels=state_list_updated, square = True, vmax=1, vmin=0)


    pl.yticks(fontname = 'Arial', fontsize = 18)
    pl.xticks(fontname = 'Arial', fontsize = 18)

    pl.axvline(8.95, color='white', lw=4)
    pl.axvline(11.95, color='white', lw=4)
    pl.title(treatment)
    return(state_usages_subset_pd)

```

```python
# calculate program expression per cell state and per condition
gep_per_state_per_condition= {}
for conditions in condition_list:

    gep_per_state_per_condition[conditions] = gep_exp1(adata, conditions)
    
```
