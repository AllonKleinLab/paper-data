# What this notebook does:

This notebook plots program weights, i.e. Fig. 3E for McCauley/Kukreja et. al. 2024

# Set up

```python
import numpy as np, pandas as pd, matplotlib.pyplot as pl, scipy.sparse as ssp
import scanpy as sc,scanpy.external as sce
import sys
sys.path.append('../../helper_functions/')
import helper_functions_py3 as hf

sc.settings.verbosity = 2  # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.settings.set_figure_params(dpi=100)  # low dpi (dots per inch) yields small inline figures

import random
import scipy.stats  as ss
import statsmodels.stats.multitest as multitest
from matplotlib.patches import Rectangle
import seaborn as sn
from sklearn.neighbors import NearestNeighbors
```

# Load data

```python
# load gene program data, also called the spectra matrix:

# this is gene x programs matrix with a weight for every gene 
# for each of the gene expression programs

cnmf_dir = '../../data/cnmf/'

spectra = pd.read_csv(cnmf_dir+'program_weights.csv', sep=',', index_col=0).T
```

```python
unique_programs = ['CHIR', 'IL13','IFNA', 'IFNG-1', 'IFNG-2', 'IFNG-3', 'TGFB1-1', 'TGFB1-2']
shared_programs=['Shared-1', 'Shared-2', 'Shared-3']
```

```python
pert_genes = {}

for progs in shared_programs+unique_programs:
    pert_genes[progs] = np.array(spectra.loc[progs].sort_values(ascending=False)[:10].index)
        
```

```python
specific_prog_list = np.array((pd.DataFrame(pert_genes).iloc[:,3:].T)).flatten()
shared_prog_list = np.array((pd.DataFrame(pert_genes).iloc[:,:3].T)).flatten()

```

# Plot spectra

```python
# add top genes from shared programs, unique programs, and KRT17
spectra_perturbed = spectra.loc[shared_programs + unique_programs][np.hstack([shared_prog_list,specific_prog_list, np.array(['KRT17'])])]
```

```python
pd.Series(spectra_perturbed.columns).to_csv('../../data/csv_txt_files/spectra_heatmap_genes.csv')
```

```python
genes_to_highlight_ix = np.array([0,2,4,5,11,12,14,18,20,21,22,26,27,28,34,40,48,49,50,56,58,60,70,74,78,80,81,93,97,100,101,106,107,110])
genes_to_highlight = spectra_perturbed.columns[genes_to_highlight_ix]
```

```python
pert_program_labels = ['Shared-1','Shared-2','Shared-3','CHIR','IL13','IFNA',
                       'IFNG-1','IFNG-2','IFNG-3','TGFB1-1','TGFB1-2']
pl.figure(figsize = (6,4))
sn.heatmap(spectra_perturbed*100,
           cmap = 'Reds' , vmax = 1.2, rasterized = True, cbar=[])
bli = pl.yticks(np.arange(spectra_perturbed.shape[0])+0.5,pert_program_labels,
                rotation=0, fontstyle='italic', fontsize = 13)
bla = pl.xticks( genes_to_highlight_ix+0.5, labels = [], fontstyle='italic') 
pl.axvline(x=10, color = 'lightgrey')
pl.axvline(x=20, color = 'lightgrey')
pl.axvline(x=30, color = 'lightgrey')
pl.axvline(x=40, color = 'lightgrey')
pl.axvline(x=50, color = 'lightgrey')
pl.axvline(x=60, color = 'lightgrey')
pl.axvline(x=70, color = 'lightgrey')
pl.axvline(x=80, color = 'lightgrey')
pl.axvline(x=90, color = 'lightgrey')
pl.axvline(x=100, color = 'lightgrey')
pl.grid(False)

bla = pl.xticks( [110+0.5], labels = ['KRT17'], fontstyle='italic') 


pl.savefig('../../figures/spectra_loadings_top10_genes.pdf',bbox_inches='tight', dpi = 300)
```
