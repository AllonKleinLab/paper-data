```python
#Code for loading SMARTseq3 data
from scipy.sparse import coo_matrix, vstack

def load_annotated_text(file_data, delim='\t', read_row_labels=False, read_column_labels=False, transpose=False, chunk_size=2000):
    '''Load text file as scipy.sparse.csc_matrix, returning column and/or row labels if desired.
    Loads rows in chunks to ease memory demands.
    '''

    X_data = []
    X_row = []
    X_col = []
    
    ncol = None
    nrow = 0

    row_labels = []
    column_labels = []

    E_chunks = []

    for row_ix, dat in enumerate(file_data):
        if type(dat) == bytes:
            dat = dat.decode('utf-8')
            
        dat = dat.strip('\n').split(delim)

        if read_column_labels and row_ix == 0:
            if read_column_labels:
                column_labels = dat[1:]
            else:
                column_labels = dat
        else:
            if read_row_labels:
                row_labels.append(dat[0])
                rowdat = dat[1:]
            else:
                rowdat = dat[0:]
            if ncol is None:
                ncol = len(rowdat)
            else:
                if len(rowdat) != ncol:
                    return 'ERROR: Line {} has {} columns. Previous line(s) had {}'.format(row_ix, len(rowdat), ncol)
            rowdat = np.array(list(map(float, rowdat)))
            col_ix = np.nonzero(rowdat)[0]
            X_col.extend(col_ix)
            X_row.extend([nrow] * len(col_ix))
            X_data.extend(rowdat[col_ix])
            nrow += 1

            if chunk_size is not None:
                if nrow % chunk_size == 0:
                    E_chunks.append(coo_matrix((X_data, (X_row, X_col)), dtype=float, shape=(nrow, ncol)))
                    X_data = []
                    X_row = []
                    X_col = []
                    nrow = 0

    if nrow > 0:
        E_chunks.append(coo_matrix((X_data, (X_row, X_col)), dtype=float, shape=(nrow, ncol)))

    E = vstack(E_chunks)
    if transpose: 
        E = E.T
    
    return E.tocsc(), np.array(row_labels), np.array(column_labels)

def file_opener(filename):
    '''Open file and return a file object, automatically decompressing zip and gzip 

    Arguments
    - filename : str
        Name of input file

    Returns
    - outData : file object
        (Decompressed) file data

    '''
    if filename.endswith('.gz'):
        fileData = open(filename, 'rb')
        import gzip
        outData = gzip.GzipFile(fileobj = fileData, mode = 'rb')
    elif filename.endswith('.zip'):
        fileData = open(filename, 'rb')
        import zipfile
        zipData = zipfile.ZipFile(fileData, 'r')
        fnClean = filename.strip('/').split('/')[-1][:-4]
        outData = zipData.open(fnClean)
    else:
        outData = open(filename, 'r')
    return outData
```

### SMARTseq3xpress data can be downloaded from https://www.ebi.ac.uk/biostudies/arrayexpress/studies/E-MTAB-11452

```python
#Load SMARTseq3xpress data 
X, barcodes, genes = load_annotated_text(
    file_opener('data/PBMCs/Smart3express/PBMCs.allruns.umicounts_intronexon.txt'),
    delim='\t',
    read_row_labels=True,
    read_column_labels=True
)
D['SMART'] = sc.AnnData(X.T)
#D['SMART'].obs_names = genes
D['SMART'].var_names = barcodes
D['SMART'].var_names_make_unique()
del X, barcodes, genes

meta = pd.read_csv('data/PBMCs/Smart3express/PBMCs.allruns.barcode_annotation.txt', sep = '\t')
D['SMART'].obs = meta
```

```python
#Filter the freshly loaded mtx
sc.pp.filter_cells(D['SMART'], min_counts=3000)
sc.pp.filter_cells(D['10X_website'], min_counts=3000)
```

```python
#Append N_gene and N_counts
for s in ['10X_website', 'SMART']:
    D[s].obs['n_genes'] = (D[s].X.A>0).sum(1)
    D[s].obs['n_counts'] = D[s].X.A.sum(1)
```

```python
#Visualize raw genes vs UMIs plots for each cell in the dataset
fig,ax = plt.subplots(1, 4, figsize=(6,2), dpi = 200)

#color lib
cols = ['teal', 'black', 'gray', 'darkslateblue']

#iterate over samples and plot
for iS, s in enumerate(['CAGEs', '10X_STARsolo', '10X_website', 'SMART']):
     ax[iS].scatter(D[s].obs['n_counts'], D[s].obs['n_genes'], marker = 'o', facecolors = 'none', edgecolors=cols[iS],s=1, label =s, alpha = 0.3)


for ax in ax:
    ax.set_xlim(0, 60e3)
    ax.set_ylim(0, 10e3)
    ax.set_xticks([0, 20e3, 40e3, 60e3])
    ax.set_yticks([0, 2.5e3, 5e3, 7.5e3, 10e3])
    ax.set_xticklabels(['0', '20', '40', '60'])
    ax.set_yticklabels(['0', '2.5', '5', '7.5', '10'])
    ax.set_xlabel(r'UMIs/cell ($x10^{3}$)')
    ax.set_ylabel(r'Genes/cell ($x10^{3}$)')
fig.tight_layout()
```
