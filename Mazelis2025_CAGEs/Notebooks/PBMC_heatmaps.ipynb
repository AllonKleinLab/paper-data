## Coarse-grain transcriptomes

### Coarse-graining code

```python
def aggregate_adata(adata, group_by='cell_type_cg'):
    # Calculate mean expression for each group
    X_new = None
    if isinstance(adata.X, sparse.spmatrix):
        X_new = sparse.vstack([
            sparse.csr_matrix(adata[adata.obs[group_by] == g].X.mean(axis=0)).reshape(1, -1)
            for g in adata.obs[group_by].unique()
        ])
    else:
        X_new = np.vstack([
            np.array(adata[adata.obs[group_by] == g].X.mean(axis=0)).reshape(1, -1)
            for g in adata.obs[group_by].unique()
        ])
    
    # Create new AnnData object
    adata_new = AnnData(
        X=X_new,
        var=adata.var.copy(),
        obs=pd.DataFrame(index=adata.obs[group_by].unique())
    )
    
    # Add cell type as a column in obs
    adata_new.obs[group_by] = adata_new.obs.index
    
    return adata_new


def aggregate_to_dataframe(adata, group_by='cell_type_cg'):
    # Calculate mean expression for each group
    if isinstance(adata.X, sparse.spmatrix):
        X_new = sparse.vstack([
            sparse.csr_matrix(adata[adata.obs[group_by] == g].X.mean(axis=0)).reshape(1, -1)
            for g in adata.obs[group_by].unique()
        ])
        # Convert sparse matrix to dense array
        X_new = X_new.toarray()
    else:
        X_new = np.vstack([
            np.array(adata[adata.obs[group_by] == g].X.mean(axis=0)).reshape(1, -1)
            for g in adata.obs[group_by].unique()
        ])
    
    # Create DataFrame using the gene names as columns and cell types as index
    df = pd.DataFrame(
        data=X_new,
        columns=adata.var_names,
        index=adata.obs[group_by].unique()
    )
    
    return df
```

### To major cell type

```python

# Create aggregated AnnData
#full_adata.obs = adata.obs
df_major_type = aggregate_to_dataframe(full_adata, group_by='coarse_labels')

# Print shapes to verify
print(f"Original shape: {full_adata.shape}")
print(f"Aggregated shape: {df_major_type.shape}")
```

```python
# Order the cell types
ordered_labels = ['HPC','Platelet','pDC', 'cDC1', 'cDC2', 'Mono. CD14','Mono. CD16', 'B','Plasmoblast', 'T', 'NK']

# Reorder the dataframe
df_major_type = df_major_type.loc[ordered_labels]
```

```python
# Normalize to max, and remove genes with zero expression across all cells

# Find genes with non-zero expression
gene_max = df_major_type.max(axis=0) 
non_zero_genes = gene_max > 0  
print(f"Number of genes before filtering: {df_major_type.shape[1]}")
print(f"Number of genes with zero expression: {sum(~non_zero_genes)}")

# Filter the dataframe
df_major_type = df_major_type.loc[:, non_zero_genes]
gene_max = gene_max[non_zero_genes]
print(f"Number of genes after filtering: {df_major_type.shape[1]}")

# Now proceed with normalization on the filtered data
df_major_type_norm = df_major_type.copy()
df_major_type_norm = df_major_type_norm / gene_max
```

## Plots

```python
marker_genes = {
    'HPC': ['CD34', 'KIT', 'MYB', 'GATA2', 'PBX1'],
    'Platelet': ['PF4', 'GP1BA', 'ITGA2B','TUBB1','PPBP'],
    'pDC': ['LILRA4', 'CLEC4C', 'TCF4',  'IL3RA','IRF8'],
    'cDC1': ['HLA-DRA','XCR1', 'CLEC9A', 'BATF3'],
    'cDC2': ['CD1C', 'FCER1A', 'CLEC10A'],
    'Mono. CD14': ['LYZ','CD14', 'FCN1', 'S100A8'], #,'CSF3R'
    'Mono. CD16': ['FCGR3A', 'MS4A7', 'CDKN1C','CSF1R'], #'PECAM1',
    'B': ['CD19', 'MS4A1', 'IGHD', 'FCER2','CD80'],
    'Plasmoblast': ['CD27','IRF4', 'PRDM1',  'IGHG1'],
    'T': ['CD3E','CD8A','IL7R','TCF7', 'ICOS'],
    'NK': [ 'GZMA','XCL1','CD247','NKG7','IFNG'],
}

# Filter the dictionary to exclude genes not in the database:
valid_genes = {}

# Iterate through each cell type and its markers
for cell_type, genes in marker_genes.items():
    valid_genes[cell_type] = []
    for gene in genes:
        if gene in adata.var_names:
            valid_genes[cell_type].append(gene)
        else:
            print(f"Gene not found in dataset for {cell_type}: {gene}")
            
# Remove any cell types that have no valid genes left
valid_genes = {k: v for k, v in valid_genes.items() if v}

# Print summary of filtering
print("\nSummary:")
for cell_type, genes in valid_genes.items():
    print(f"{cell_type}: {len(genes)} valid genes remaining")



```

```python
from matplotlib.collections import LineCollection

# Create the ordered genes list and track positions
ordered_genes = []
ordered_groups = []
current_position = 0
for cell_type, genes in valid_genes.items():
    ordered_genes.extend(genes)
    current_position += len(genes)
    ordered_groups.append((cell_type, current_position))

# Create the plot
n_genes = len(ordered_genes)
fig_width = max(5, n_genes * 0.12)
fig, ax = plt.subplots(figsize=(fig_width, 3), dpi=200)
ax = sns.heatmap(
    df_major_type_norm.loc[:,ordered_genes],
    cmap='viridis',
    yticklabels=df_major_type_norm.index,
    xticklabels=True,
    cbar_kws={'label': 'Normalized Expression'},
    vmin=0,
    vmax=1,
    center=None
)

# Add vertical lines after every gene group
for _, pos in ordered_groups[:-1]:  # Exclude the last position
    plt.axvline(x=pos, color='white', linewidth=2)

plt.xticks(
    np.arange(len(ordered_genes)) + 0.5,
    ordered_genes,
    rotation=90,
    ha='center',
    fontsize=6
)
plt.yticks(rotation=0, fontsize=7)

# Add group labels above the plot
group_centers = []
prev_pos = 0
for cell_type, pos in ordered_groups:
    center = prev_pos + (pos - prev_pos)/2
    group_centers.append((center, cell_type))
    prev_pos = pos

# Add the group labels
ax2 = ax.twiny()
ax2.set_xlim(ax.get_xlim())
ax2.set_xticks([x for x, _ in group_centers])
ax2.set_xticklabels([label for _, label in group_centers], rotation=45, ha='left',fontsize=8)
ax2.tick_params(axis='x', length=0)


# Adjust layout to prevent text truncation
plt.tight_layout()
# Add extra padding at the top for the group labels
plt.subplots_adjust(top=0.85)  # Adjust this value as needed (0.85 means 15% padding at top)

#plt.savefig('fig_major_celltypes.pdf',bbox_inches='tight',  # This ensures nothing is cut off
#            dpi=300,             # Higher DPI for better quality
#            format='pdf')
#plt.tight_layout()
plt.show()
plt.close()
```

# Now make a plot zooming into NKT cell subsets

```python
#Load TNK subest data
adata_TNK = sc.read_h5ad('data/PBMCs/PBMC_TNK.h5ad')
```

```python
# Create a new AnnData object with raw dimensions from the layer
TNK_unscaled = sc.AnnData(
    X=adata_TNK.layers['unscaled'].copy(),
    obs=adata_TNK.obs,
    var=adata_TNK.var
)
```

```python
#Log scale the data
adata_log = sc.pp.log1p(TNK_unscaled, base=10,copy=True)
adata_log.obs['cell_type_sub'] = TNK_unscaled.obs['cell_type_sub']

# Perform ranking for each cell type
sc.tl.rank_genes_groups(
    adata_log,
    groupby='cell_type_sub',
    method='wilcoxon',  # You can also use 't-test' or other methods
    key_added='cell_type_enriched'
)

# Create a dictionary to store enriched genes for each cell type
enriched_genes = {}

# For each cell type, get significantly enriched genes
for cell_type in TNK_unscaled.obs['cell_type_sub'].unique():
    # Get the ranked genes for this cell type
    genes = sc.get.rank_genes_groups_df(
        adata_log, 
        key='cell_type_enriched', 
        group=cell_type
    )
    
    # Filter for significant genes (adjust these thresholds as needed)
    significant_genes = genes[
        (genes['pvals_adj'] < 0.05) &  # adjusted p-value threshold
        (genes['logfoldchanges'] > 1)   # log fold change threshold
    ]
    
    # Store the gene names
    enriched_genes[cell_type] = significant_genes['names'].tolist()
    
    # Print summary for this cell type
    print(f"\n{cell_type}:")
    print(f"Found {len(enriched_genes[cell_type])} enriched genes")
    print("Top 30 genes:", enriched_genes[cell_type][:30])

# Optional: Save the full results
results_df = sc.get.rank_genes_groups_df(
    adata_log,
    key='cell_type_enriched',
    group=list(TNK_unscaled.obs['cell_type_sub'].unique())
)
```

```python

# Create aggregated AnnData
df_cell_type = aggregate_to_dataframe(TNK_unscaled, group_by='cell_type_sub')

# Print shapes to verify
print(f"Original shape: {TNK_unscaled.shape}")
print(f"Aggregated shape: {df_cell_type.shape}")
```

```python
# Order the cell types
ordered_labels = ['NK$_{CD56-lo}$', 'NK$_{CD56-hi}$', 'NK$_{Mem}$',
                  'CD8 T$_{Naive/Mem}$', 
                  'CD8 T$_{cyt}$-1', 
                  'CD8 T$_{cyt}$-2',
                  'CD8 T$_{cyt}$-3', 
                  'CD8 T$_{cyt}$-4',
                  'CD8 T$_{Naive/Mem}^{Helios+}$', 
                  'CD4 T$_{Naive/Mem}$', 
                  'CD4 T$_{eff}$-1', 
                  'CD4 T$_{eff}$-2',  
                  'CD4 T$_{eff}$-3', 
                  'T$_{MAIT}$', 
                  'T$_{cyc.}$']
# Reorder the dataframe
df_cell_type = df_cell_type.loc[ordered_labels]

```

```python
# Create a dictionary to store top 30 genes for each cell type
top30_genes = {}
for cell_type in adata_log.obs['cell_type_sub'].unique():
    genes = sc.get.rank_genes_groups_df(
       adata_log, 
       key='cell_type_enriched', 
       group=cell_type
   )
   
    significant_genes = genes[
       (genes['pvals_adj'] < 0.05) &
       (genes['logfoldchanges'] > 1)
    ].sort_values('logfoldchanges', ascending=False)
   
    top30_genes[cell_type] = significant_genes['names'].head(30).tolist()

seen_genes = set()
unique_top_genes = {}

for cell_type in adata_log.obs['cell_type_sub'].unique():
    gene_list = top30_genes[cell_type]
    unique_genes = [gene for gene in gene_list if gene not in seen_genes]
    seen_genes.update(unique_genes)
    unique_top_genes[cell_type] = unique_genes

top30_genes = unique_top_genes

top30_genes = {cell_type: top30_genes[cell_type] for cell_type in df_cell_type_norm.index}
```

```python

# Create a new ordered dictionary
ordered_top_genes = {label: top30_genes[label] for label in ordered_labels}
top30_genes = ordered_top_genes
```

```python
# Create the ordered genes list and track positions
ordered_genes = []
ordered_groups = []
current_position = 0

for cell_type, genes in top30_genes.items():
    ordered_genes.extend(genes)
    current_position += len(genes)
    ordered_groups.append((cell_type, current_position))

# Create the plot
n_genes = len(ordered_genes)
fig_width = max(5, n_genes * 0.02)
fig, ax = plt.subplots(figsize=(fig_width, 2), dpi=200)
ax = sns.heatmap(
    df_cell_type_norm.loc[:,ordered_genes],
    cmap='viridis',
    yticklabels=df_cell_type_norm.index,
    xticklabels=False,
    cbar_kws={'label': 'Normalized Expression'},
    vmin=0,
    vmax=1,
    center=None,
    linewidths=0  # Remove fine white lines
)

#Add white lines between groups
for _, pos in ordered_groups[:-1]:
     plt.axvline(x=pos, color='white', linewidth=0.5)

plt.yticks(rotation=0, fontsize=7)

# Add group labels above plot
group_centers = []
prev_pos = 0
for cell_type, pos in ordered_groups:
    center = prev_pos + (pos - prev_pos)/2
    group_centers.append((center, cell_type))
    prev_pos = pos

ax2 = ax.twiny()
ax2.set_xlim(ax.get_xlim())
ax2.set_xticks([x for x, _ in group_centers])
ax2.set_xticklabels([label for _, label in group_centers], rotation=45, ha='left', fontsize=7)
ax2.tick_params(axis='x', length=0)

plt.tight_layout()
plt.subplots_adjust(top=0.85)
#plt.savefig('fig_NK-T_cell_subsets_top30.pdf', bbox_inches='tight', dpi=300, format='pdf')
plt.show()
```
