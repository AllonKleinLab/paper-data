```python
def aggregate_adata(adata, group_by='cell_type_cg'):
    # Calculate mean expression for each group
    X_new = None
    if isinstance(adata.X, sparse.spmatrix):
        X_new = sparse.vstack([
            sparse.csr_matrix(adata[adata.obs[group_by] == g].X.mean(axis=0)).reshape(1, -1)
            for g in adata.obs[group_by].unique()
        ])
    else:
        X_new = np.vstack([
            np.array(adata[adata.obs[group_by] == g].X.mean(axis=0)).reshape(1, -1)
            for g in adata.obs[group_by].unique()
        ])
    
    # Create new AnnData object
    adata_new = AnnData(
        X=X_new,
        var=adata.var.copy(),
        obs=pd.DataFrame(index=adata.obs[group_by].unique())
    )
    
    # Add cell type as a column in obs
    adata_new.obs[group_by] = adata_new.obs.index
    
    return adata_new


def aggregate_to_dataframe(adata, group_by='cell_type_cg'):
    # Calculate mean expression for each group
    if isinstance(adata.X, sparse.spmatrix):
        X_new = sparse.vstack([
            sparse.csr_matrix(adata[adata.obs[group_by] == g].X.mean(axis=0)).reshape(1, -1)
            for g in adata.obs[group_by].unique()
        ])
        # Convert sparse matrix to dense array
        X_new = X_new.toarray()
    else:
        X_new = np.vstack([
            np.array(adata[adata.obs[group_by] == g].X.mean(axis=0)).reshape(1, -1)
            for g in adata.obs[group_by].unique()
        ])
    
    # Create DataFrame using the gene names as columns and cell types as index
    df = pd.DataFrame(
        data=X_new,
        columns=adata.var_names,
        index=adata.obs[group_by].unique()
    )
    
    return df
```

```python
adata_log = sc.pp.log1p(adata_unscaled, base=10,copy=True)
adata_log.obs['cell_type_sub'] = adata_TNK.obs['cell_type_sub']

# Perform ranking for each cell type
sc.tl.rank_genes_groups(
    adata_log,
    groupby='cell_type_sub',
    method='wilcoxon',  # You can also use 't-test' or other methods
    key_added='cell_type_enriched'
)

# Create a dictionary to store enriched genes for each cell type
enriched_genes = {}

# For each cell type, get significantly enriched genes
for cell_type in adata_TNK.obs['cell_type_sub'].unique():
    # Get the ranked genes for this cell type
    genes = sc.get.rank_genes_groups_df(
        adata_log, 
        key='cell_type_enriched', 
        group=cell_type
    )
    
    # Filter for significant genes (adjust these thresholds as needed)
    significant_genes = genes[
        (genes['pvals_adj'] < 0.05) &  # adjusted p-value threshold
        (genes['logfoldchanges'] > 1)   # log fold change threshold
    ]
    
    # Store the gene names
    enriched_genes[cell_type] = significant_genes['names'].tolist()
    
    # Print summary for this cell type
    print(f"\n{cell_type}:")
    print(f"Found {len(enriched_genes[cell_type])} enriched genes")
    print("Top 10 genes:", enriched_genes[cell_type][:10])

# Optional: Save the full results
results_df = sc.get.rank_genes_groups_df(
    adata_log,
    key='cell_type_enriched',
    group=list(adata_TNK.obs['cell_type_sub'].unique())
)
```

```python
# Create a dictionary to store top 5 genes for each cell type
top_genes = {}

# For each cell type, get top 5 genes
for cell_type in adata_log.obs['cell_type_sub'].unique():
    # Get the ranked genes for this cell type
    genes = sc.get.rank_genes_groups_df(
        adata_log, 
        key='cell_type_enriched', 
        group=cell_type
    )
    
    # Sort by logfoldchange and get top 5 significant genes
    significant_genes = genes[
        (genes['pvals_adj'] < 0.05) &
        (genes['logfoldchanges'] > 1)
    ].sort_values('logfoldchanges', ascending=False)
    
    # Store top 5 gene names
    top_genes[cell_type] = significant_genes['names'].head(5).tolist()

# Keep track of genes we've seen
seen_genes = set()

# Create a new dictionary with only unique genes
unique_top_genes = {}

# Go through cell types in order
for cell_type in adata_log.obs['cell_type_sub'].unique():
    gene_list = top_genes[cell_type]
    
    # Filter out any genes we've seen before
    unique_genes = [gene for gene in gene_list if gene not in seen_genes]
    
    # Update our seen genes set
    seen_genes.update(unique_genes)
    
    # Store the unique genes for this cell type
    unique_top_genes[cell_type] = unique_genes

# Replace original dictionary
top_genes = unique_top_genes
```

```python
def update_marker_genes(marker_genes, cell_type=None, gene_list=None, remove=False):
   """
   Update marker genes dictionary: add or remove genes.
   If remove=True, cell_type is ignored and genes are removed from all lists where found.
   """

   if isinstance(gene_list, str):
       new_genes = [gene_list]

       
   if remove:
       for gene in gene_list:
           found = False
           for ct, genes in marker_genes.items():
               if gene in genes:
                   marker_genes[ct].remove(gene)
                   print(f"Removed {gene} from {ct}")
                   found = True
           if not found:
               print(f"Gene {gene} not found in any cell type list")
       
   else:
       if isinstance(gene_list, str):
           new_genes = [new_genes]
       
       if cell_type not in marker_genes:
           marker_genes[cell_type] = []
       
       for gene in gene_list:
           found = False
           for existing_type, genes in marker_genes.items():
               if gene in genes:
                   print(f"Gene {gene} is already a marker for {existing_type}. Not added.")
                   found = True
                   break
           
           if not found:
               marker_genes[cell_type].append(gene)
```

**Recent reference defining NK subsets by scRNA-Seq:**
* Reference: https://www.nature.com/articles/s41590-024-01883-0
* The resolution of NK subsets in our data is far sharper than in this paper

**NK 3:**
* These are canonical CD56-dim NK cells
* Key reference: https://pubmed.ncbi.nlm.nih.gov/11698225/
* Considered more mature, based on:
    - CD56dim (NCAM1) lower
    - Express KIRs (e.g. KIR2DL1)
    - Express cytotoxic genes (GZMB, PRF1)
    - Express higher levels of the chemokine receptor CX3CR1
    - Lower Sellectin (SELL)
    - Higher CD57 (B3GAT1)
    - CD16+ (but not brightest) cells

**NK 2:**
* This profile matches quite well with CD56bright NK cells
* Key reference: https://pubmed.ncbi.nlm.nih.gov/11698225/
* Properties:
    * Very high NCAM1
    * Also express: SELL
    * Lower expression of KIRs (KIR2DL1)

**NK 1:**
* This population appears to be what's called "memory" NK cells
    * https://www.nature.com/articles/nri.2015.9
    * https://pubmed.ncbi.nlm.nih.gov/21825173/ - (NKG2C-high, where this is KLRC2)
* Has mature NK cell markers, shared with NK3 (GZMB, PRF1, FCGR3A, CX3CR1, B3GAT1/CD57)
* Also like NK3, it has lower CD56 (NCAM1) and lower NCR1 relative to NK2
* Expresses KLRC2 (NKG2C) - this is a defining feature of adaptive NK cells ??
* Also express: ZBTB38

```python
# Add another category of common genes
marker_genes = {'T/NK': ['CD3D','NCAM1','NCR1']}
marker_genes.update(top_genes)

# Add some marker genes that are well-known
update_marker_genes(marker_genes, 'NK 1', ['IFNG','KLRC2','ZBTB38']) 
update_marker_genes(marker_genes, 'NK 2', ['SELL'])
update_marker_genes(marker_genes, 'NK 3', ['GZMB','PRF1','FCGR3A','CX3CR1'])  #,'ZNF516','NFIL3'
update_marker_genes(marker_genes, 'CD4 T$_{eff}$-2', ['ISM1','ANK3'])  #,'ZNF516','NFIL3'
update_marker_genes(marker_genes, 'CD4 T$_{eff}$-3', ['FAAHP1'])  #,'ZNF516','NFIL3'


# Remove some markers:
update_marker_genes(marker_genes,gene_list=[
    'CHSY3', 'LINGO2', 'EPB41L4A','AL732406.1','ADAMTS1','KLRA1P','AC093865.1','IATPR','AC068775.1','LINC01644',
    'AC010967.1','AC139720.1', 'AL353660.1','LINC01727','LINC02446','AL162493.1','LEF1-AS1','LMNA'],
                    remove=True)
```

```python
# Rename the NK subsets, and re-order
# New copy of the data matrix:
df_cell_type_norm_renamed = df_cell_type_norm.copy()

# First rename the rows:
renaming_dict = {'NK 2': 'NK$_{CD56-hi}$','NK 1': 'NK$_{Mem}$','NK 3': 'NK$_{CD56-lo}$'}
df_cell_type_norm_renamed.rename(index=renaming_dict, inplace=True)

# Now re-order the rows
new_order = [
    'NK$_{CD56-lo}$', 'NK$_{CD56-hi}$', 'NK$_{Mem}$', 
    'CD8 T$_{Naive/Mem}$', 'CD8 T$_{cyt}$-1', 'CD8 T$_{cyt}$-2', 'CD8 T$_{cyt}$-3', 'CD8 T$_{cyt}$-4', 'CD8 T$_{Naive/Mem}^{Helios+}$', 
    'CD4 T$_{Naive/Mem}$', 'CD4 T$_{eff}$-1', 'CD4 T$_{eff}$-2', 'CD4 T$_{eff}$-3', 
    'T$_{MAIT}$', 'T$_{cyc.}$']

df_cell_type_norm_renamed = df_cell_type_norm_renamed.loc[new_order]

# Now rename the marker gene dictionary:
marker_genes_renamed = marker_genes.copy()
marker_genes_renamed.update({ 'NK$_{CD56-hi}$': marker_genes_renamed.pop('NK 2')})
marker_genes_renamed.update({ 'NK$_{Mem}$': marker_genes_renamed.pop('NK 1')})
marker_genes_renamed.update({ 'NK$_{CD56-lo}$': marker_genes_renamed.pop('NK 3')})

# # Now re-order the marker gene dictionary:
new_order = ['T/NK', 
    'NK$_{CD56-lo}$', 'NK$_{CD56-hi}$', 'NK$_{Mem}$', 
    'CD8 T$_{Naive/Mem}$', 'CD8 T$_{cyt}$-1', 'CD8 T$_{cyt}$-2', 'CD8 T$_{cyt}$-3', 'CD8 T$_{cyt}$-4', 'CD8 T$_{Naive/Mem}^{Helios+}$', 
    'CD4 T$_{Naive/Mem}$', 'CD4 T$_{eff}$-1', 'CD4 T$_{eff}$-2', 'CD4 T$_{eff}$-3', 
    'T$_{MAIT}$', 'T$_{cyc.}$']
marker_genes_renamed = {k: marker_genes_renamed[k] for k in new_order}

```

```python
# Create the ordered genes list and track positions
ordered_genes = []
ordered_groups = []
current_position = 0

# Use the ordered marker_genes dictionary 
for cell_type, genes in marker_genes_renamed.items():
    ordered_genes.extend(genes)
    current_position += len(genes)
    ordered_groups.append((cell_type, current_position))

# Create the plot
n_genes = len(ordered_genes)
fig_width = max(5, n_genes * 0.12)
fig, ax = plt.subplots(figsize=(fig_width, 3), dpi=200)
ax = sns.heatmap(
    df_cell_type_norm_renamed.loc[:,ordered_genes],
    cmap='viridis',
    yticklabels=df_cell_type_norm_renamed.index,
    xticklabels=True,
    cbar_kws={'label': 'Normalized Expression'},
    vmin=0,
    vmax=1,
    center=None
)

# Rest of the plotting code remains the same
for _, pos in ordered_groups[:-1]:
    plt.axvline(x=pos, color='white', linewidth=2)

plt.xticks(
    np.arange(len(ordered_genes)) + 0.5,
    ordered_genes,
    rotation=90,
    ha='center',
    fontsize=6
)
plt.yticks(rotation=0, fontsize=7)

# Add group labels above the plot
group_centers = []
prev_pos = 0
for cell_type, pos in ordered_groups:
    center = prev_pos + (pos - prev_pos)/2
    group_centers.append((center, cell_type))
    prev_pos = pos

# Add the group labels
ax2 = ax.twiny()
ax2.set_xlim(ax.get_xlim())
ax2.set_xticks([x for x, _ in group_centers])
ax2.set_xticklabels([label for _, label in group_centers], rotation=45, ha='left', fontsize=8)
ax2.tick_params(axis='x', length=0)

plt.tight_layout()
plt.subplots_adjust(top=0.85)
#plt.savefig('fig_NK-T_cell_subsets.pdf', bbox_inches='tight', dpi=300, format='pdf')
plt.show()
```
