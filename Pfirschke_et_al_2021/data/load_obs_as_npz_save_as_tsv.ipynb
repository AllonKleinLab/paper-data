# Npz to tsv

GEO does not accept .npz files. Load cell annotations ("obs") as npz and save as tsv.

```python
import pandas as pd
import numpy as np
```

```python
def load_df(filename,encoding=u'ASCII'):
    """you may want to specify encoding='latin1'
    when loading python 2 pickle with python 3.
    https://stackoverflow.com/questions/28218466/unpickling-a-python-2-object-with-python-3
    """
    with np.load(filename,encoding=encoding) as f:
        obj = pd.DataFrame(**f)
    return obj
```

```python
obs = load_df('obs_info_27563x32_201025_14h44.npz')
print(obs.shape)
obs.head()
```

```python
obs.to_csv('obs_info_27563x32_201025_14h44.tsv',sep='\t')
```
